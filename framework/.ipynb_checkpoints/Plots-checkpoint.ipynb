### Example realization of PP

```python



from tick.base import TimeFunction

from tick.plot import plot_point_process
from tick.hawkes import SimuInhomogeneousPoisson
run_time = 30

T = np.arange((run_time ) * 5, dtype=float) / 5
Y = np.array([max(0.01,2*np.sin(i*0.2+5)) for i in T])

tf = TimeFunction((T, Y), dt=0.001)
# We define a 1 dimensional inhomogeneous Poisson process with the
# intensity function seen above
in_poi = SimuInhomogeneousPoisson([tf], end_time=run_time, verbose=False)
# We activate intensity tracking and launch simulation
in_poi.track_intensity(0.1)
in_poi.simulate()
fig, ax = plt.subplots(1, 1, figsize=(10, 4))
# We plot the resultifig, ax = plt.subplots(1, 1, figsize=(10, 4))ng inhomogeneous Poisson process with its
# intensity and its ticks over time
plot_point_process(in_poi,ax=ax)
ax.set_ylabel("$\lambda(t)$", fontsize=18)
ax.set_title("Realization of a point process")
```

### Plot realization of BuSca process

```python

def plot_gb(timestamps, mu, alpha, beta, tmax):
    intensity = mu
    intensities = []
    t_idx = 0
    event = timestamps[t_idx]
    plt.plot(timestamps[0],intensity,'o', color = "#1F77B4", markersize = 6.5, alpha=0.9)
    for t in np.arange(0,tmax,0.01):
        if t > event:
            
            if t_idx > 0:
                delta_t = timestamps[t_idx] - timestamps[t_idx-1]
                intensity = mu + alpha/(beta+delta_t)
                plt.plot(timestamps[t_idx]+1e-2,intensity+1e-3,'o', color = "#1F77B4", markersize = 6.5, alpha=0.9)
            if t_idx+1 < len(timestamps):
                t_idx += 1
                event = timestamps[t_idx]
        intensities.append(intensity)
    plt.plot(np.arange(0,tmax,0.01), intensities,'-', color = "#1F77B4")
    plt.xlabel(r'$t$', size = 16)
    plt.ylabel(r'$\lambda(t|\mathcal{H}(t))$', size = 16)
    plt.title('BuSca Process', size = 15)
    plt.show()
    
d = 1
mu = 0.1
beta = 2
Mu = np.ones((d,1)) * mu
Alpha = np.ones((d,d)) / d
Beta = np.ones((d,d)) * beta

simulation_time = 80
sim = simulate.GrangeBuscaSimulator(mu_rates = Mu, Alpha_ba = Alpha)
timestamps = sim.simulate(simulation_time)
plot_gb(timestamps[0], mu, 1, beta, tmax=simulation_time)
    
```

```python
def get_metricval(groundtruth, adjlist):
    metric_result = []
    for adj in adjlist:
        try:
            metric_result.append(metric(groundtruth, adj))
        except:
            pass
    metric_result = sorted(metric_result)
    return metric_result

def get_prob_less_and_lq(metric_result):
    num_samples = len(metric_result)

    P_less = [0]
    P_leq = [1/num_samples]
    last_repeated_sq = [0]
    for i in range(1,num_samples):
        if metric_result[i] != metric_result[i-1]:
            last_repeated_sq = [i]
            P_less.append(i/num_samples)
            P_leq.append((i+1)/num_samples)
        else:
            P_less.append(P_less[-1])
            P_leq.append((i+1)/num_samples)
            for j in last_repeated_sq:
                P_leq[j] = (i+1)/num_samples
            last_repeated_sq.append(i) 
    
    return P_less, P_leq

def get_e_var_vn(metric_result, n):
    num_samples = len(metric_result)
    expected_vn = 0
    var_vn = 0
    prev_value = -1e10
    for i in range(num_samples):
        if metric_result[i] != prev_value:
            expected_vn += metric_result[i]*(P_leq[i]**n - P_less[i]**n)
            var_vn += (metric_result[i]**2)*(P_leq[i]**n - P_less[i]**n)
        prev_value = metric_result[i]
    var_vn = var_vn - expected_vn**2
    
    return expected_vn, var_vn
    
file_list = ["sx-mathoverflow", "sx-askubuntu", "sx-superuser", "CollegeMsg", "email-Eu-core-temporal", "wiki-talk-temporal","memetracker_2009-01"]
models = ["HkEM", "ADM4", "SG", "Cumulant", "ExpKern", "Granger"]
metrics = [rankcorr, apk, ndcg, roc_auc]
metricnames = ["RankCorr","AP@n","NDCG","AUC"]


for file in file_list:
    fig, axlist = plt.subplots(2,2,figsize=(14,8))#,facecolor='w', dpi= 80)
    axi = 0
    for metric in metrics:
        for model in models:
            print(np.argmax())
            ax = axlist[axi//2][axi%2]

            metric_result = get_metricval(groundtruth[file], adjacency[file][model])

            P_less, P_leq = get_prob_less_and_lq(metric_result)

            Y = []
            VARY = []
            hp_tries = range(1,60,2)
            for n in hp_tries:

                expected_vn, var_vn = get_e_var_vn(metric_result, n)
                assert var_vn >= 0
                VARY.append(var_vn)
                Y.append(expected_vn)

            Y = np.array(Y)
            VARY = np.array(VARY)

            ax.plot(hp_tries, Y, label=model)
            ax.fill_between(hp_tries, Y-VARY, Y+VARY, alpha = 0.4)

        ax.set_title(metricnames[axi])
        axi += 1
    ax.legend(loc='center right',  bbox_to_anchor=(0.8, 0., 0.5, 0.5))
    fig.suptitle(file,size=13)
    plt.show()
```
