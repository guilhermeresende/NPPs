### Simulate Hawkes Plaw

```python
d=10
mu = 0.02 * np.ones(d) 
Alpha = np.zeros((d, d))
Beta = np.ones((d, d))
for i in range(d):
    for j in range(d):
        if (i==j) or (i%2==0 and (j-i)==1) or (i%2==1 and (i-j)==1):
            Alpha[i][j]=0.3
            Beta[i][j]=0.5

plt.imshow(Alpha)

kernels=[[hk.HawkesKernelExp(Alpha[i][j],Beta[i][j]) for j in range(len(Beta[i]))] for i in range(len(Alpha))]
h = hk.SimuHawkes(force_simulation=True,kernels=kernels, baseline=list(mu), max_jumps=10000)
dt = 0.01
h.track_intensity(dt)
h.simulate()
timestamps=h.timestamps
sim_time=h.simulation_time
print([len(i) for i in timestamps])
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import ShuffleSplit

class GrangerExt(BaseEstimator, TransformerMixin):

    def __init__(self, mtype="ADM4", decay=None,C=None,lasso_nuclear_ratio=None,tol=None,kernel_support=None,kernel_size=None,integration_support=None,penalty=None,solver=None,step=None):
        self.mtype=mtype
        if(self.mtype=="EM"):
            self.kernel_support = 10. if (kernel_support is None) else kernel_support
            self.kernel_size = 10 if (kernel_size is None) else kernel_size
            self.tol=1e-05 if (tol is None) else tol
        elif(self.mtype=="ADM4"):
            self.decay = 1. if (decay is None) else decay
            self.C = 1000. if (C is None) else C
            self.lasso_nuclear_ratio = 0.5 if (lasso_nuclear_ratio is None) else lasso_nuclear_ratio
            self.tol=1e-05 if (tol is None) else tol
        elif(self.mtype=="HC"):
            self.integration_support = 20.0 if (integration_support is None) else integration_support
            self.C = 1000. if (C is None) else C
            self.penalty='none' if (penalty is None) else penalty
            self.solver='adagrad' if (solver is None) else solver
            self.step=1e2 if (step is None) else step
            self.tol=1e-05 if (tol is None) else tol
            


    # the arguments are ignored anyway, so we make them optional
    def fit(self, timestamps):
        if(self.mtype=="EM"):
            model= hk.HawkesEM(kernel_support=self.kernel_support,kernel_size=self.kernel_size, tol=self.tol, max_iter=300)
            model.fit(timestamps)
        elif(self.mtype=="ADM4"):
            model= hk.HawkesADM4(decay=self.decay,C=self.C,lasso_nuclear_ratio=self.lasso_nuclear_ratio, tol=self.tol, max_iter=300)
            model.fit(timestamps)
        elif(self.mtype=="HC"):
            try:
                model= hk.HawkesCumulantMatching(integration_support=self.integration_support,C=self.C,penalty=self.penalty, solver=self.solver, step=self.step,tol=self.tol, max_iter=300)
                model.fit(timestamps)
            except:
                self.model=-1
        self.model=model

    def transform(self, timestamps):
        return model

    def fit_transform(self, timestamps):
        if(self.mtype=="ADM4"):
            model= hk.HawkesADM4(decay=self.decay,C=self.C,lasso_nuclear_ratio=self.lasso_nuclear_ratio, tol=self.tol, max_iter=300)
        #model.fit(timestampst)
        return model

    def score(self,timestamps):
        if(self.mtype=="EM"):
            return self.model.score()
        elif(self.mtype=="ADM4"):
            return self.model.score()
        elif(self.mtype=="HC"):
            try:
                return -self.model.objective(self.model.adjacency)
            except:
                return -1e20
```

```python
def kernel_values(model, i, j, t_begin, t_end):
    dx = 0.1
    t_values = np.arange(t_begin, t_end + dx, dx)

    y_values = model.get_kernel_values(i, j, t_values)
    
    return y_values.sum()*dx

def integral_hawkes( timestamps,proc_a, n_a):    
    process = timestamps[proc_a]
    interval_end = n_a
    interval_begin = interval_end-1
    interval_len = process[interval_end]-process[interval_begin]
    value = 0
    
    dx = 0.1 
    for proc_b in range(len(timestamps)):
        i = 0
        T_i = timestamps[proc_b][i]

        while(T_i < process[interval_begin] and i<(len(timestamps[proc_b]) - 1)):
            value += kernel_values(model, proc_a, proc_b, process[interval_begin] - T_i, process[interval_end] - T_i)   
            i += 1
            T_i = timestamps[proc_b][i]
            
        while(T_i < process[interval_end] and i < (len(timestamps[proc_b]) - 1)):
            value += kernel_values(model, proc_a, proc_b, 0, process[interval_end] - T_i)   
            i += 1
            T_i = timestamps[proc_b][i]
            
    return value + model.baseline[proc_a]*interval_len
    
def get_zj_hawkes(model, timestamps):
   
    zjlist = []
    
    for proc_a in range(len(timestamps)):
        print(proc_a,"1")
        #for n_a in range(2, len(timestamps[proc_a])):
        #    zjlist.append(integral_hawkes(model, timestamps, proc_a, n_a))
        #args = [(timestamps,proc_a,n_a) for n_a in range(2,len(timestamps[proc_a]))]
        args = [(timestamps,proc_a,n_a) for n_a in random.sample(range(2,len(timestamps[proc_a])),30)]
        for timestamps, proc_a, n_a in args:
            zjlist.append(integral_hawkes(timestamps, proc_a, n_a))
            
        print(proc_a)
        #with Pool(2) as p:
        #    zjlist += (p.starmap(integral_hawkes,args))

    return zjlist

def ks_plot(zjlist): 
    zjlist = np.array(zjlist)
    n = len(zjlist)
    zk = 1-np.exp(-zjlist) #transform into uniform
    sorted_zk = sorted(zk)
    y_values = np.array([(k-0.5)/n for k in range(1,n+1)])
    plt.plot(sorted_zk, y_values, c='green')
    plt.plot(y_values, y_values+1.36/(n**0.5), linestyle='--', c='k', alpha=0.75)
    plt.plot(y_values, y_values-1.36/(n**0.5), linestyle='--', c='k', alpha=0.75)
    plt.show()

#decay,C,lasso_nuclear_ratio,tol=select_hyper_ADM4(timestamps, n_iter = 30)
decay,C,lasso_nuclear_ratio,tol = 0.21179255964235422,  3381.0117904988124, 0.25842109659208978, 3.9888607932821158e-05 
model_adm4 = hk.HawkesADM4(decay=decay,C=C,lasso_nuclear_ratio=lasso_nuclear_ratio, tol=tol, max_iter=300,verbose=True)
model_adm4.fit(timestamps)
model = model_adm4
np.seterr(under='warn')
zjlist = get_zj_hawkes(model_adm4, timestamps)
ks_plot(zjlist)
ss.probplot(zjlist, dist = ss.expon, plot = plt)
plt.show()
print(ss.kstest(zjlist,'expon'))
```

```python
def binary_search(x,l,low,high):
    mid=(low+high)//2
    if ((mid+1)>=len(l)):
        return mid
    elif (mid<0):
        return -1
    elif  (l[mid]<=x and l[mid+1]>x):
        return mid    
    else:
        if l[mid]>x:
            return binary_search(x,l,low,mid-1)
        
        else:
            return binary_search(x,l,mid+1,high)


def calc_delta(time,timelist):
    index=binary_search(time,timelist,0,len(timelist)-1)
    if index==-1:
        return False
    return time - timelist[index]


def calc_value(interval_len,delta_list,mu,Beta,Alpha,process_num):
    value=mu[process_num]
    for j in range(len(delta_list)):
        if delta_list[j]>0:
            value+=Alpha[j][process_num]/(Beta[j][process_num]+delta_list[j])
    return (value*interval_len)

def integral_granger(timestamps,proc_a,interval_end,mu,Beta,Alpha):    
    process=timestamps[proc_a]
    interval_len=process[interval_end]-process[interval_end-1]
    delta_list=[]                
    for proc_b in range(len(timestamps)):
        if(proc_b==proc_a):
            delta=process[interval_end-1]-process[interval_end-2]
        else:
            delta=calc_delta(process[interval_end-1],timestamps[proc_b])
        if(delta==False):
            delta_list.append(0)
        else:
            delta_list.append(delta)
    if len(delta_list)==len(timestamps):
        value=calc_value(interval_len,delta_list,mu,Beta,Alpha,process_num=proc_a)
        return value
    
def get_zj_granger(timestamps,mu,Alpha,Beta):
   
    zjlist=[]
    for proc_a in range(len(timestamps)): 
        process=timestamps[proc_a]

        args=[(timestamps,proc_a,i,mu,Beta,Alpha) for i in range(2,len(process))]
        with Pool(10) as p:
            zjlist+=(p.starmap(integral_granger,args))

    return zjlist

def ks_plot(zjlist): 
    zjlist=np.array(zjlist)
    n=len(zjlist)
    zk=1-np.exp(-zjlist) #transform into uniform
    sorted_zk = sorted(zk)
    y_values = np.array([(k-0.5)/n for k in range(1,n+1)])
    plt.plot(sorted_zk, y_values, c='green')
    plt.plot(y_values, y_values+1.36/(n**0.5), linestyle='--', c='k')
    plt.plot(y_values, y_values-1.36/(n**0.5), linestyle='--', c='k')
    plt.show()

np.seterr(under='warn')
granger_model = GrangerBusca(alpha_prior=1.0/len(timestamps), num_iter=300,beta_strategy=1)
granger_model.fit(timestamps)
Alpha_t = normalize(granger_model.Alpha_.toarray().T, "l1")
mu_t = granger_model.mu_
Beta_t = np.ones(shape=(len(Alpha_t),len(Alpha_t)))
zjlist_granger=get_zj_granger(timestamps,mu_t,Alpha_t,Beta_t)
ss.probplot(zjlist_granger,dist=ss.expon,plot=plt)
plt.show()
ks_plot(zjlist_granger)
print(ss.kstest(zjlist_granger,'expon'))
```

```python
def ks_diff(simulations,ground_truth):
    ks_sum=0
    for simulated in simulations:
        deltas=get_deltas(simulated)
        deltas_gt=get_deltas(ground_truth)
        for p1 in range(len(deltas)):
            try:
                ks_sum+=ss.ks_2samp(deltas[p1],deltas_gt[p1])[0]
            except:
                print("erro ks")
    return ks_sum/(len(ground_truth)*len(simulations))
def diff_corr(simulations,ground_truth):
    corr_sum=0
    for simulated in simulations:
        deltas=get_deltas(simulated)
        deltas_gt=get_deltas(ground_truth)
        for p1 in range(len(deltas)):
            try:
                deltas_minus1=deltas[p1][:-1]
                deltas_p=deltas[p1][1:]            
                corr=ss.pearsonr(deltas_p,deltas_minus1)[0]

                deltas_minus1_gt=deltas_gt[p1][:-1]
                deltas_p_gt=deltas_gt[p1][1:]            
                corr_gt=ss.pearsonr(deltas_p_gt,deltas_minus1_gt)[0]

                corr_sum+=abs(corr-corr_gt)
            except:
                print("erro corr")
           
    return corr_sum/(len(ground_truth)*len(simulations))
        


def diff_var(simulations,ground_truth):
    var_diff=0
    for simulated in simulations:
        deltas=get_deltas(simulated)
        deltas_gt=get_deltas(ground_truth)
        for p1 in range(len(deltas)):
            try:
                var_=np.var(deltas[p1])
                var_gt=np.var(deltas_gt[p1])
                var_diff+=abs(var_-var_gt)
            except:
                print("erro var")
    return var_diff/(len(ground_truth)*len(simulations)) 

def diff_coef_var(simulations,ground_truth):
    coef_diff=0
    for simulated in simulations:
        deltas=get_deltas(simulated)
        deltas_gt=get_deltas(ground_truth)
        for p1 in range(len(deltas)):
            try:
                coef_=np.std(deltas[p1])/np.mean(deltas[p1])
                coef_gt=np.std(deltas_gt[p1])/np.mean(deltas_gt[p1])

                coef_diff+=abs(coef_-coef_gt)
            except:
                print("erro coef")
    return var_diff/(len(ground_truth)*len(simulations)) 
```
