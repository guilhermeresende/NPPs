```python
import numpy as np
import matplotlib.pyplot as plt
import tick.hawkes as hk
from scipy import integrate
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
d = 1
intensity = np.ones(d)*0.3
n_events = 1000
h = hk.SimuPoissonProcess(intensities = intensity, max_jumps = n_events, verbose = False)
h.simulate()
timestamps = h.timestamps
counting = [np.arange(len(timestamps[i])) for i in range(len(timestamps))]
for i in range(len(timestamps)):
    plt.plot(timestamps[i], counting[i], label='%d' % i)
timestamps = timestamps[0]
plt.show()
    
def func(x, lambdap, beta):
    return np.exp(-lambdap*x)/(x+beta)
    
def distance(lambdap, beta, timestamps):
    #print("s",lambdap,beta,"s")
    T = max(timestamps)
    n = len(timestamps)
    d = T*lambdap -n*np.log(lambdap)
    integ = integrate.quad(lambda x: func(x,lambdap,beta),0,np.inf)[0]
    d -= len(timestamps)*(2*integ + np.log(beta))
    return d

print(len(timestamps)/max(timestamps))


result = []
dt = 0.1
X = np.arange(1e-6,10,dt)
lambdap = len(timestamps)/max(timestamps)

minv = float("inf")
mini = 0
lambdap = len(timestamps)/max(timestamps)

for beta in X:
    d = distance(lambdap,beta,timestamps)
    result.append(d)
    if d < minv:
        minv = d
        mini = beta
print("min beta",mini)
plt.plot(X,result)
plt.show()


minv = float("inf")
mini = 0
beta = 1
result = []
for lambdap in X:
    d = distance(lambdap,beta,timestamps)
    result.append(d)
    if d < minv:
        minv = d
        mini = lambdap
print("min lambda",mini)
plt.plot(X,result,'r')
plt.show()

```
