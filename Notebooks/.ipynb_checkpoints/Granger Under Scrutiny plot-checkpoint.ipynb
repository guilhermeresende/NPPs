```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import scipy.stats as ss
import matplotlib.lines as mlines
from matplotlib.patches import Rectangle

import sys
sys.path.append("../framework/")
from pp_metrics import *

import matplotlib as mpl
mpl.style.use('default')
```

```python
table = []
table_up = []
table_low = []
null_model = []
models = []
datasets = []
modelnames = set('GB', 'HC', 'MLE-SGLP', 'ADM4', 'HkEM', '')


with open("../framework/results_all_metrics.txt","r") as f:
    dataset = f.readline().strip("\n")
    while(dataset!=""):
        model = ""
        while (model in modelnames):
            model = f.readline().strip("\n")
            metrics = f.readline().strip("\n").split("\t")
            metrics_val = [float(metrics[i].replace(",","").replace("(","").replace(")","").split(" ")[0]) for i in range(len(metricnames))]
            metrics_val2 = [float(metrics[i].replace(",","").replace("(","").replace(")","").split(" ")[1]) for i in range(len(metricnames))]
            metrics_val3 = [float(metrics[i].replace(",","").replace("(","").replace(")","").split(" ")[2]) for i in range(len(metricnames))]
            #if model!="Null model" and not(model=="SG" and dataset=="memetracker_2009-01"):
            if model!="Null model":
                models.append(model)
                datasets.append(dataset if not("memetracker" in dataset) else "Memetracker")
                table.append(np.array(metrics_val))
                table_up.append(np.array(metrics_val2))
                table_low.append(np.array(metrics_val3))
            else:
                null_model.append(np.array([metrics_val,metrics_val2,metrics_val3]))

        dataset = f.readline().strip("\n")

table = np.array(table)
table_up = np.array(table_up)
table_low = np.array(table_low)
null_model = np.array(null_model)
print(set(datasets))
print(set(models))
```

```python
#dataset_colors = {"CollegeMsg": '#B8213A', "sx-askubuntu": '#5A39BF', "sx-mathoverflow": '#3AA1A8', \
#                   "sx-superuser": '#61C244', "email-Eu-core-temporal": '#B89D42',  "Memetracker": 'k'}
datasetnames = ['Memetracker', 'sx-superuser', 'sx-askubuntu', 'sx-mathoverflow', 'wiki-talk-temporal', 'email-Eu-core-temporal', 'CollegeMsg'][::-1]
dataset_colors = {"Memetracker": 'k', "wiki-talk-temporal": '#B8671D',"email-Eu-core-temporal": '#B89D42',\
                "sx-superuser": '#61C244', "sx-mathoverflow": '#3AA1A8', "sx-askubuntu": '#5A39BF',"CollegeMsg": '#B8213A'}
model_shapes = {'GB':'v', 'HC':'s', 'MLE-SGLP':'D', 'ADM4':'o', 'HkEM':'^'}
metricnames = ["RankCorr","P@n","NDCG","AUC","NRMSE", "AP@n"]
metricorder = {"RankCorr": 0,"P@n": 1,"NDCG": 3,"AUC":4,"NRMSE":5, "AP@n": 2}
num_models = len(model_shapes)
def newline(ax,p1, p2, color='black'):
    l = mlines.Line2D([p1[0],p2[0]], [p1[1],p2[1]], color=color, alpha=0.7,zorder=1,lw=2.3, solid_capstyle='round')
    ax.add_line(l)
    return l

# Figure and Axes
fig, axlist = plt.subplots(2,3,figsize=(6*3,6.5*2),facecolor='w', dpi= 80)
#print(len(axlist),len(axlist[0]))
for m in range(len(metricnames)):
    m_idx = metricorder[metricnames[m]]
    ax = axlist[m_idx//3][m_idx%3]
    values = table[:,m]
    boot_up = table_up[:,m]
    boot_low = table_low[:,m]
    null_range = null_model[:,:,m]

    maxv = max(max(values),max(boot_up))
    minv = min(min(values),min(boot_low))
    # Vertical Lines
    #for k in np.arange(-1.0,1,0.1):
    #    ax.vlines(x=k, ymin=0, ymax=maxy, color='black', alpha=0.7, linewidth=1, linestyles='dotted'
    
    gap = 3
    maxy = len(values) + 7*gap
    for i in range(0,len(values)):
        yh = i + (i//5)*gap
        #if(i%5==0):
        #    yh = i + (i//5)
        #print(i,yh)
        #shaded area
        if (i%num_models == 0):
            height = num_models
            y_low = yh
            #if(i== len(values)-num_models):
            #    height = num_models+2
            #elif(i==0):
            #    height = num_models + 0.35
            #    y_low = -0.35

            x_low = minv-0.05
            rectrange = null_range[int(i/num_models)][2] - x_low
            if (metricnames[m]=="NRMSE"):
                x_low = null_range[int(i/num_models)][1]
                rectrange = (maxv +0.05) - x_low
            rect = Rectangle((x_low,y_low),zorder=3, width = rectrange, height=height, color='#939393',alpha=0.5,linewidth=0)
            ax.add_patch(rect)
        #markers
        ax.scatter(y=yh, x = values[i], zorder=2, s=60, color=dataset_colors[datasets[i]], \
                   edgecolor='k', marker = model_shapes[models[i]],alpha=0.8)
        #line
        newline(ax,[boot_up[i],yh],[boot_low[i],yh], color=dataset_colors[datasets[i]])
        #text
        if (i%5==0 and i>0):
            text = ax.text(x=minv+(maxv-minv)/2,y=yh-1, s=datasets[i-1], ha="center", va="top", color="k", size = 15.2)
            ax.hlines(y=yh-gap,xmin=minv-0.05,xmax=maxv*1.15, color='black', alpha=0.7, linewidth=1, linestyles='dotted')
            ax.hlines(y=yh-8.1,xmin=minv-0.05,xmax=maxv*1.15,zorder=0, color='black', alpha=0.7, linewidth=1, linestyles='dotted')
    i = len(values)
    yh = i + (i//5)*gap
    text = ax.text(x=minv+(maxv-minv)/2,y=yh-1, s=datasets[-1], ha="center", va="top", color="k", size = 15.2)
    ax.hlines(y=yh-gap,xmin=minv-0.05,xmax=maxv*1.15, color='black', alpha=0.7, linewidth=1, linestyles='dotted')
    ax.hlines(y=yh-8.1,xmin=minv-0.05,xmax=maxv*1.15, color='black', alpha=0.7, linewidth=1, linestyles='dotted')

    ax.hlines(y=maxy-0.05,xmin=minv-0.05,xmax=maxv*1.15, color='black', alpha=0.7, linewidth=1, linestyles='dotted')
    # Decoration
    ax.set_facecolor('w')
    ax.set_title(metricnames[m], fontdict={'size':23},loc='left')
    ax.set_xticks(np.arange(np.round(minv,decimals=1),1,0.1))
    ax.set(xlim=(minv-0.05,maxv*1.05), ylim=(-0.4, maxy))
    ax.set_yticks(np.round(np.arange(0,len(values),2.5)))
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    ax.get_yaxis().set_visible(False)


    

    #Legend
    legend=[]
    for m in model_shapes:
        legend.append(mlines.Line2D([], [], color='#4f4f4f', linestyle='None',marker=model_shapes[m],markersize=10, label=m))
    #for d in dataset_colors:
    #    legend.append(mlines.Line2D([0], [0], color=dataset_colors[d], lw=8, label=d))
    ax.legend(handles=legend,loc='upper left', frameon=False, ncol=5, bbox_to_anchor=(0.27, 1.06),columnspacing=0.,handletextpad=0.,borderpad=0.,fontsize=11.5)
    #(handles=legend,loc='upper left', ncol=5, frameon=False, bbox_to_anchor=(0.4, 1.3),fontsize=8,columnspacing=0.1,borderpad=0.,labelspacing=0., handletextpad=0.)

plt.tight_layout()
plt.savefig("results.png")
plt.show()
```
