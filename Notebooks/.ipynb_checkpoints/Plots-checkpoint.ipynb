```python
%matplotlib inline
import sys
sys.path.append("../../granger-busca")
import numpy as np
from gb import simulate, GrangerBusca
import tick
import tick.hawkes as hk
import matplotlib.pyplot as plt
from multiprocessing import Pool
from time import time
import scipy.stats as ss
from tick.plot import plot_point_process
```

```python



from tick.base import TimeFunction

from tick.plot import plot_point_process
from tick.hawkes import SimuInhomogeneousPoisson
run_time = 30

T = np.arange((run_time ) * 5, dtype=float) / 5
Y = np.array([max(0.01,2*np.sin(i*0.2+5)) for i in T])

tf = TimeFunction((T, Y), dt=0.001)
# We define a 1 dimensional inhomogeneous Poisson process with the
# intensity function seen above
in_poi = SimuInhomogeneousPoisson([tf], end_time=run_time, verbose=False)
# We activate intensity tracking and launch simulation
in_poi.track_intensity(0.1)
in_poi.simulate()
fig, ax = plt.subplots(1, 1, figsize=(10, 4))
# We plot the resultifig, ax = plt.subplots(1, 1, figsize=(10, 4))ng inhomogeneous Poisson process with its
# intensity and its ticks over time
plot_point_process(in_poi,ax=ax)
ax.set_ylabel("$\lambda(t)$", fontsize=18)
ax.set_title("Realization of a point process")
```

```python
import matplotlib.pyplot as plt
import numpy as np

from tick.base import TimeFunction
from tick.hawkes import SimuHawkes, HawkesKernelExp, HawkesKernelTimeFunc
from tick.plot import plot_point_process

t_values = np.array(np.arange(0,20.,0.01), dtype=float)
kernels = []
for i in range(1,2):
    y_values = np.sin(t_values*8*i) / 4 
    #plt.plot(t_values,y_values);plt.show()
    tf1 = TimeFunction([t_values, y_values],
                       inter_mode=TimeFunction.InterConstRight, dt=0.01)
    kernel = HawkesKernelTimeFunc(tf1)
    kernels.append(kernel)


hawkes = SimuHawkes(
    kernels=[np.random.permutation(kernels) for i in kernels],
    baseline=[1]*len(kernels))

run_time = 20
dt = 0.01
hawkes.track_intensity(dt)
hawkes.end_time = run_time
hawkes.threshold_negative_intensity()
hawkes.simulate()

fig, ax = plt.subplots(hawkes.n_nodes, 1, figsize=(14, 8))
plot_point_process(hawkes, t_max=30, ax=ax)

plt.show()
```
